```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('apple_products.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
highest_rated = data.sort_values(by="Star Rating", ascending=False).head(10)
```

```python
highest_rated
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=highest_rated, x='Product Name', y='Number Of Ratings', hue='Star Rating', dodge=False)
plt.xticks(rotation=25, ha='right')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=highest_rated, x='Product Name', y='Number Of Reviews', hue='Star Rating', dodge=False)
plt.xticks(rotation=-25, ha='left')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=data, x='Number Of Ratings',  y='Sale Price',size='Discount Percentage', sizes=(20,200))
sns.regplot(data=data, x='Number Of Ratings',  y='Sale Price')

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=data, x='Number Of Ratings',  y='Discount Percentage',size='Sale Price', sizes=(20, 200))
sns.regplot(data=data, x='Number Of Ratings',  y='Discount Percentage')

plt.xticks(rotation=45)
plt.show()
```
