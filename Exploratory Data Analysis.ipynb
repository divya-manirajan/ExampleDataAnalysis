```python
quality_counts = data['quality'].value_counts()
```

```python
quality_counts
```

```python
quality_counts
```

```python
quality_counts.values
```

```python
plt.figure(figsize=(15, 5))

sns.barplot(x=quality_counts.index, y=quality_counts.values, palette='viridis')

plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')

plt.show()
```

```python
fig, ax = plt.subplots(7, 2, figsize=(20,30))

row = 0
col = 0

for column in data.columns: 
    sns.histplot(x=data[column], kde='True', ax=ax[row,col])
    col += 1
    if col > 1:
        col = 0
        row += 1

plt.show()
```

```python
plt.figure(figsize=(15, 5))

sns.swarmplot(data=data, x='quality', y='alcohol', palette='rocket')
plt.title('Swarm Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')

plt.show()
```
